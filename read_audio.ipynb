## Reading and plotting audio data

Audio singnals are complex continuous-valued waves. In order to save a digial version of it, we should convert our sample data into number presentation. If speech is commonly sampled at *44100 Hz*, so we store a value every *1*/*44100* seconds.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
# read the input file
sampling_freq, audio = wavfile.read('input_read.wav')
```

```python
# out audio parameters
print ('Shape: '+ str(audio.shape))
print ('Datastype: ' + str(audio.dtype))
print ('Duration: ' + str(round(audio.shape[0] / float(sampling_freq), 3)) + ' second')
```

```python
# audio signal is stored as 16-big sined integer data.
# normalize these values
audio = audio / (2.**15)
```

```python
# extract first 30 values for plotting
audio = audio[:30]
```

```python
# build time histogram
x_values = np.arange(0, len(audio),1) / float(sampling_freq)
```

```python
# converts the units to seconds
x_values *= 1000
```

```python
# plot
plt.plot(x_values, audio, color='black')
plt.xlabel('Times (ms)')
plt.ylabel('Amplitude')
plt.title('Audio signal')
plt.show()
```
