## Extracting frequeny domain features

At the second stage we convet a signal into the freqency domain. In monst modern speech recognitoon freqeency-domain features are used as key component. After convert a signal into a freq domain, it's requered to convert it into a useable form. **Mel Frequency Cepstral Coefficients (MFCC)** is a good way to do that. *MFCC* takes the power spectrum of a signal and then uses a combination of filter banks and disrete cosinetransform to extract pattern of phones or features.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile 
from python_speech_features import mfcc, logfbank
```

```python
# read input sound file
sampling_freq, audio = wavfile.read("input_freq.wav")
```

```python
# extract MFCC and filter bank features
mfcc_features = mfcc(audio, sampling_freq)
filterbank_features = logfbank(audio, sampling_freq)
```

```python
# print the parameters
print ('MFCC:\nNumber of windows =' + str(mfcc_features.shape[0]))
print ('Length of each feature =' + str(mfcc_features.shape[1]))
print ('\nFilter bank:\nNumber of windows =' + str(filterbank_features.shape[0]))
print ('Length of each feature =' + str(filterbank_features.shape[1]))
```

```python
# plot hte MFCC features
mfcc_features = mfcc_features.T # transfrom matrix so the time domain is horizontal
plt.matshow(mfcc_features)
plt.title('MFCC')
plt.show()
```

```python
# plot hte Filter bank features

filterbank_features = filterbank_features.T  # transfrom matrix so the time domain is horizontal
plt.matshow(filterbank_features)
plt.title('Filter bank')
plt.show()
```
